# Train and Test split
Here due to imbalance between both target variables, I will divide the **Train Set** and **Test Set** in such a way that:
1. The **Train Set** should have balanced rows for both `spam` and `ham` categories.
2. The **Test Set** can have imbalanced rows, since we just need to get the results.
#### So Final values to assign
1. **Train Set** (`train-data.csv`):
    - 1280 rows of `spam`.
    - 1280 rows of `ham`.
    - **Total: 2560 rows.**

2. **Test Set** (`test-data.csv`):
    - 372 rows of `spam`.
    - 4516 rows of `ham`.
    - **Total: 4888 rows.**

```python
# Random Shuffle
df_shuffled = df_comb.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled
```

```python
df_spam = df_shuffled[df_shuffled["v1"] == "spam"].reset_index(drop=True)
df_spam
```

```python
df_ham = df_shuffled[df_shuffled["v1"] == "ham"].reset_index(drop=True)
df_ham
```

### Train Dataset

```python
# Setting Up Training Dataset
df_spam_1280 = df_spam.iloc[:1280]
df_ham_1280 = df_ham.iloc[:1280]

df_train = pd.concat([df_spam_1280, df_ham_1280], axis=0, ignore_index=True)
df_train = df_train.sample(frac=1, random_state=42).reset_index(drop=True)
df_train = pd.get_dummies(df_train, columns=["v1"], dtype=int)
df_train.columns = ["Text", "Ham", "Spam"]
df_train
```

```python
df_train.to_csv("Prepared-Data/train-data.csv", index=False)
```

### Test Dataset

```python
# Setting up test dataset as per planning
df_spam_372 = df_spam.iloc[1280:]
df_ham_3236 = df_ham.iloc[1280:]

df_test = pd.concat([df_spam, df_ham_3236], axis=0, ignore_index=True)
df_test = df_test.sample(frac=1, random_state=42).reset_index(drop=True)
df_test = pd.get_dummies(df_test, columns=["v1"], dtype=int)
df_test.columns = ["Target", "Ham", "Spam"]
df_test
```

```python
df_test.to_csv("Prepared-Data/test-data.csv", index=False)
```
